### 새로운 열 r을 생성하고 반지름을 계산해보자

## ggplot : 온도 구간을 세분화 하여 시각화

### 목표; 모든 온도 구간에 대해 각각 색을 다르게 할당하여 그려보자

`*` 우리가 봐야하는 변수는 온도, 아이스크림 판매량, 소아마비

`*` 온도가 유사한 지역을 색으로 묶으면 3차원 플랏이 가능함

#### 자료를 데이터 프레임으로 정리해보자

## 구간 나누기 방법 1

```python
def cut(x): # 이거보다 더 좋은 방법이 있을 것 같긴 한데요..
    if x<0: 
        y='Temp: <0'
    elif x<5: 
        y='Temp: 0~5'
    elif x<10: 
        y='Temp: 5~10'
    elif x<15: 
        y='Temp: 10~15'
    elif x<20:
        y='Temp: 15~20'
    elif x<25: 
        y='Temp: 20~25'
    else: 
        y='Temp: >30'
    return y
```

```python
df.assign(temp2 = list(map(cut,df.temp)))
```

### 시각화 시작

#### 온도를 색으로 구분

```python
fig = ggplot(data = df.assign(temp2 = list(map(cut,df.temp))))
```

```python
p1 = geom_point(aes(x='icecream',y='disease',colour='temp2'),alpha=0.5)
```

```python
fig + p1
```

#### 추세선 추가

```python
l1 = geom_smooth(aes(x='icecream',y='disease',colour='temp2'))
```

```python
fig + p1 + l1
```

#### 해석

`-` 각 온도별로 추세선은 거의 기울기가 0. -> 온도가 비슷한 구간별로 묶어서 보니까 상관관계가 없다는 것이다.

`-` 아이스크림 판매량과 소아마비의 corr은 유의미해 보였지만, 온도를 통제하였을 경유 아이스크림 판매량과 소아마비의 partial corr는 유의미해 보이지 않았다

`-` 온도와 같은 변수를 은닉변수라고 한다. 뜻은 이름에서 보듯이 은닉된 변수

### 만약 아이스크림이랑 소아마비가 진짜 관련이 있다면?

```python
np.random.seed(1)
eps= np.random.normal(size= len(temp), scale= 10)
icecream = 20 + 2* temp + eps
```

`-` 수영장이 원인이 아니라 진짜 아이스크림을 먹고 소아마비에 걸린 상황이라면

```python
np.random.seed(2)
eps = np.random.normal(size= len(temp),scale =2)
disease = 30 + 0*temp + 0.15* icecream + eps
```

```python
df2 = pd.DataFrame({'temp':temp,'icecream':icecream,'disease':disease})
df2.assign(temp2=list(map(cut,df2.temp)))
```

```python
ggplot(data= df2.assign(temp2= list(map(cut,df2.temp)))) +\
geom_point(aes(x='icecream',y='disease',colour='temp2'),alpha=0.2) +\
geom_smooth(aes(x='icecream',y='disease',colour='temp2'))
```
