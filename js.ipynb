```python
import pandas as pd
import numpy as np
```

```python
_dflst = pd.read_html('https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%A0%80%EC%B6%9C%EC%82%B0')
_df1 = _dflst[2] 
_df2 = _dflst[4] 
```

```python
df2 = _df2.set_index('지역/연도[6]')\
.applymap(lambda x: None if x == '-' else float(x))\
.applymap(lambda x: x*1000 if x<1000 else x)\
.drop('전국')\
.reset_index()\
.melt(id_vars='지역/연도[6]')\
.set_axis(['지역','연도','출생아수'],axis=1)
df2 # 강의에서 이용한 자료 입니다

df2.query('지역 == "서울"')
```

```python
df2.groupby('연도')[['출생아수']].agg(np.sum).reset_index().set_axis(['연도','총 출생아수'],axis=1)
#이 데이터 프레임의 우측 '총 출생아수'부분이 첫 데이터 프레임의 오른쪽 열에 나타나게 구현을 하고 싶습니다.
```

`-` 방법1

```python
_total =df2.groupby('연도')[['출생아수']].agg(np.sum).reset_index().set_axis(['연도','총 출생아수'],axis=1).loc[:,'총 출생아수']
list(_total)
```

```python
df2.query('지역 == "서울"').assign(total = list(_total))
```

`-` 방법2

```python
df2.query('지역 == "서울"')\
.assign(total = 
       df2.
        groupby('연도')[['출생아수']].
        agg(np.sum).reset_index().
        set_axis(['연도','총 출생아수'],axis=1).
        loc[:,'총 출생아수'].pipe(list)
       )
```
