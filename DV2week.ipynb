#### 우리의 목적

- 너무 히스토그램이 몰려서 컴퓨터의 기능을 못살린다
- 그렇다면 저 히스토그램을 펼쳐보자

```python
img2_red = cv2.equalizeHist(img[...,0])
```

```python
plt.hist(img2_red.reshape(-1))
```

```python
_fig = plt.hist(img2_red.reshape(-1), bins = 255, range = [0,255])
```

```python
_fig2 = plt.hist(img[:,:,0].reshape(-1), bins = 255,range = [0,255]) #화소가 쏠려있다
```
