## 4번

```python
df = pd.read_csv('https://raw.githubusercontent.com/guebin/DV2021/master/_notebooks/2021-10-25-FIFA22_official_data.csv')
df
```

```python
df.info()
```

```python
df2 = df.drop(columns=['Loaned From', 'Marking']).dropna()
```

```python
df2.assign(Potential2 = df2['Potential'] - df2['Overall'])
```

```python
def f(x):
    if x>26 : 
        y='21~26'
    elif x>22:
        y='20~21'
    elif x>20:
        y='20<'
    else:
        y='<100' 
    return y
```

```python
_df= df2.assign(Potential2 = df2['Potential'] - df2['Overall'])\
.assign(grouped_Age = lambda df: list(map(f,df2.Age)))
```

```python
df3= _df.groupby(by="grouped_Age").agg({'Age':np.mean}).reset_index()\
.rename(columns={'Age':'Age_x'}).merge(_df)
```

```python
df3
```

```python
ggplot(data =df3) +\
geom_point(aes(x='grouped_Age',y='Potential2',color='grouped_Age'),position='jitter',size=0.1,alpha=0.5)+\
geom_boxplot(aes(x= 'grouped_Age', y='Potential2', color = 'grouped_Age'),alpha= 0.8)
```

```python
position_dict = {
    'GOALKEEPER':{'GK'},
    'DEFENDER':{'CB','RCB','LCB','RB','LB','RWB','LWB'},
    'MIDFIELDER':{'CM','RCM','LCM','CDM','RDM','LDM','CAM','RAM','LAM','RM','LM'},
    'FORWARD':{'ST','CF','RF','LF','RW','LW','RS','LS'},
    'SUB':{'SUB'},
    'RES':{'RES'}
}
position_dict
```
