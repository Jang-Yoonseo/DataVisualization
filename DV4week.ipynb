### 9월 26일 강의 

```python
np.random.seed(43052)
ϵ = np.random.randn(100)
```

```python
plt.plot(np.arange(1,101),ϵ, '--o');
```

`1` ϵt 와 ϵ(t-1)은 독립이라고 보여지는가?

독립일 것 같다. 왜 ? 정규분포에서 생성된 난수니깐?

`2` 조건을 만족하는 벡터 생성

```python
y=[]
y.append(ϵ[0])
```

```python
y
```

```python
for i in range(1,100):
    y.append(y[i-1] + ϵ[i])
```

```python
y
```

`3` t,y의 Dot-connected plot 그리기

```python
plt.plot(np.arange(1,101),y, '--o')
```

`4` 얼추 두 원소가 붙어 다니려는 경향성이 있어 보인다 (독립이 아님)

```python
np.corrcoef([np.arange(1,101),y])
```

### 과제

```python
y1 = np.random.randn(90).cumsum() 
y2 = np.random.randn(120).cumsum()
```

```python
yy = np.concatenate([y1,y2])
```

```python
hw_df = pd.DataFrame({'idx':list(range(90))+list(range(120)),'data_':yy, 'cat':['y1']*90 + ['y2']*120})
```

```python
sns.lineplot(data=hw_df,x='idx',y='data_',style='cat',hue='cat',markers=True)
```
