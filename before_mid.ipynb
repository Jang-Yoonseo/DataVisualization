```python
def f(x):
    if x>66: 
        y='66<'
    elif x>61:
        y='61~66'
    elif x>56:
        y='56~61'
    else:
        y='<56' 
    return y
```

```python
df.assign(Potential2 = df['Potential'] - df['Overall']).query('Potential2 >=1 ')\
.assign(Overall_grouped = lambda df: list(map(f, df.Overall)))
```

```python
ggplot(data = df.assign(Potential2 = df['Potential'] - df['Overall']).query('Potential2 >=1 ')\
.assign(Overall_grouped = lambda df: list(map(f, df.Overall))))\
+geom_boxplot(aes(x='Overall_grouped',y='Potential2', color= 'Overall_grouped'))
```

```python
#순서대로 나열 하고싶다.
#그룹의 평균을 구하여 그룹 전체에 부여.
#그다음에 그 축을 숫자로 하여 나열
```

```python
df.assign(Potential2 = df['Potential'] - df['Overall']).query('Potential2 >=1 ')\
.assign(Overall_grouped = lambda df: list(map(f, df.Overall))).query('Overall_grouped == "66<"')['Overall'].mean()
```

```python
def g(x):
    if x=='66<': 
        y= 71.8127687727423
    elif x=='61~66':
        y= 63.773918342474104
    elif x=='56~61':
        y= 59.155840684309005
    else:
        y= 52.87743190661479
    return y
```

```python
df.assign(Potential2 = df['Potential'] - df['Overall']).query('Potential2 >=1 ')\
.assign(Overall_grouped = lambda df: list(map(f, df.Overall)))\
.assign(Overall_X = lambda df: list(map(g,df['Overall_grouped'])))
```

```python
ggplot(data = df.assign(Potential2 = df['Potential'] - df['Overall']).query('Potential2 >=1 ')\
.assign(Overall_grouped = lambda df: list(map(f, df.Overall)))\
.assign(Overall_X = lambda df: list(map(g,df['Overall_grouped']))))\
+geom_point(aes(x='Overall', y = 'Potential2', color= 'Overall_grouped'),alpha = 0.05, position = 'jitter')\
+geom_boxplot(aes(x='Overall_X',y='Potential2',color='Overall_grouped'))
```
