### subplot 

````python
fig, axs = plt.subplots(2) #도화지 axes 2개
````

`-` 예시 1

````python
fig, axs = plt.subplots(2) #도화지에 엑시스 두개
axs[0].plot(x,y,'.b') #첫 그림
axs[1].plot(x,2*x,'r') #두 번째 그림
````

`-` 예시 2

````python
fig, axs = plt.subplots(2,2) #도화지 생성, axes(2,2)개
axs[0,0].plot(x,2*x,'--b') #첫 그림
axs[0,1].plot(x,ϵ,'.r') #두 번째 그림
axs[1,0].plot(x,y,'.r') #세 번째 그림
axs[1,1].plot(x,y,'.r') #네 번째 그림
axs[1,1].plot(x,2*x,'-b') # 네 번째의 겹치는 그림
````

## fig 와 axes의 이해: matplotlib으로 그림을 어렵게 그려내보자
