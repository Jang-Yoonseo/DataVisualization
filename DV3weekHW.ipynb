```python
import numpy as np
import matplotlib.pyplot as plt 
```

(1ë²ˆ)

```python
x,y = [1,2,3,4], [1,2,4,3] 
```

```python
fig, axs = plt.subplots(2,3)
axs[0,0].plot(x,y,'or')
axs[0,1].plot(x,y,'og')
axs[0,2].plot(x,y,'ob')
axs[1,0].plot(x,y,'--or')
axs[1,1].plot(x,y,'--og')
axs[1,2].plot(x,y,'--ob')
```

(2ë²ˆ)

```python
x,y = [1,2,3,4], [1,2,1,1]
```

```python
fig = plt.figure()
fig.add_axes([0,0,1,1])
fig.add_axes([0.5,0.5,1,1])
fig.add_axes([1,1,1,1])
ax1,ax2,ax3 = fig.axes
fig.axes[0].plot(x,y,'or')
fig.axes[1].plot(x,y,'og')
fig.axes[2].plot(x,y,'ob')
```

```python
x = np.arange(-5,5,0.1)
y1 = np.sin(x)
y2 = np.sin(2*x) + 2
y3 = np.sin(4*x) + 4 
y4 = np.sin(8*x) + 6
```

```python
plt.plot(x,y1,'--r')
plt.plot(x,y2,'--g')
plt.plot(x,y3,'--b')
plt.plot(x,y4,'--m')
```
